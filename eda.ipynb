```python
#week 2
df = df.dropna(axis=1, thresh=len(df)*0.5) # Drop columns with more than 50% missing values
```

```python
df = df.fillna(df.select_dtypes(include=['number']).median()) # Filling missing numeric values with median
```

```python
df = df.fillna(df.select_dtypes(include=['object']).mode().iloc[0]) # Filling missing categorical values with mode
```

```python
df = df.drop_duplicates() # Remove duplicate rows
```

```python
df.isnull().sum()
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')
df['End_Time'] = pd.to_datetime(df['End_Time'], errors='coerce')
df['Weather_Timestamp'] = pd.to_datetime(df['Weather_Timestamp'], errors='coerce')
# converting datetime columns in proper datetime format
```

```python
df['Hour'] = df['Start_Time'].dt.hour
df['Weekday'] = df['Start_Time'].dt.day_name()
df['Month'] = df['Start_Time'].dt.month_name()
# created new features based on Start_Time
```

```python
# now our dataset will have 3 new cloumns at the end i.e Hour, Weekday, Month
df.info()
```

```python
# distribution of accident severity levels (bar plot)
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Severity')
plt.title('Distribution of Accident Severity Levels')
plt.xlabel('Severity Level')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
#Accident frequency by Time of Day (Which hours have more accidents)
plt.figure(figsize=(12,6))
sns.countplot(x='Hour', data=df)
plt.title('Accident Frequency by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Weekday', data=df)
plt.title('Accident Frequency by Day of Week') # (Whether accidents are higher on weekdays vs weekends)
plt.xlabel('Weekday')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Month', data=df)
plt.title('Accident Frequency by Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(10,4))
df['Weather_Condition'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Weather Conditions During Accidents')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
# Weather conditions were label-encoded; therefore numeric values represent different weather categories. 
# Lower numeric values indicate more frequent weather conditions
```

```python
#most common road feature during accidents
road_features = [
    'Junction', 'Crossing', 'Stop',
    'Traffic_Signal', 'Roundabout',
    'Railway', 'No_Exit'
]

road_counts = {}

for feature in road_features:
    road_counts[feature] = df[feature].sum()

road_counts
```

```python
plt.figure(figsize=(8,4))
plt.bar(road_counts.keys(), road_counts.values())
plt.title('Accidents by Road Features')
plt.xlabel('Road Feature')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=30)
plt.show()
```

```python
# week 4 - Bivariate and Mulvariate Analysis

# We will analyze Severity vs: Weather Conditions, Visibility, Road surface condition and Traffic congestion 

#We can correlate Severity with numeric columns like Visibility, Distance, Wind_Speed, etc. (Heatmap)
cols = ['Severity', 'Visibility(mi)', 'Distance(mi)', 'Wind_Speed(mph)', 'Humidity(%)']
corr = df[cols].corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Severity vs visibility 
plt.figure(figsize=(6,4))
sns.boxplot(x='Severity', y='Visibility(mi)', data=df)
plt.title('Severity vs Visibility')
plt.show()
```

```python
# severity vs weather conditions (boxplot)
severity_weather_ct = pd.crosstab(
    df["Weather_Simple"],
    df["Severity"]
)

plt.figure(figsize=(8,4))
sns.heatmap(
    severity_weather_ct,
    annot=True,
    fmt="d",
    cmap="YlOrRd"
)

plt.title("Heatmap of Accident Severity vs Weather Condition")
plt.xlabel("Severity Level")
plt.ylabel("Weather Condition")
plt.tight_layout()
plt.show()


```

```python
# severity vs traffic congestion (Bar Plot)
df["Traffic_Congestion_Score"] = (
    df["Traffic_Signal"].astype(int) +
    df["Junction"].astype(int) +
    df["Crossing"].astype(int) +
    df["Stop"].astype(int)
)

plt.figure(figsize=(8,5))
sns.boxplot(
    x="Severity",
    y="Traffic_Congestion_Score",
    hue = "Severity",
    data=df,
    palette="Set3",
    legend=False  
)

plt.title("Accident Severity vs Traffic Congestion")
plt.xlabel("Severity Level")
plt.ylabel("Traffic Congestion Score")
plt.tight_layout()
plt.show()
```

```python
# severity vs road surface condition 
def road_surface(weather):
    weather = weather.lower()
    if 'rain' in weather:
        return 'Wet'
    elif 'snow' in weather or 'ice' in weather:
        return 'Snow/Ice'
    elif 'fog' in weather:
        return 'Foggy'
    else:
        return 'Dry'

df['Road_Surface'] = df['Weather_Condition'].astype(str).apply(road_surface)
df.groupby('Road_Surface')['Severity'].mean()
plt.figure(figsize=(6,4))
sns.barplot(
    data=df,
    x='Road_Surface',
    y='Severity',
    estimator='mean'
)
plt.title('Average Accident Severity by Road Surface Condition')
plt.xlabel('Road Surface Condition')
plt.ylabel('Average Severity')
plt.show()
```

```python
# Pair Plot (Multivariate Relationship)
sns.pairplot(df[['Severity', 'Visibility(mi)', 'Distance(mi)']])
plt.show()
```

```python
# week 5
# scatter plot 
plt.figure(figsize=(6,6))
plt.scatter(
    df['Start_Lng'],
    df['Start_Lat'],
    s=1, alpha=0.3
)
plt.title('Accident Hotspots (Latitude vs Longitude)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# density map for better visualization
plt.figure(figsize=(6,6))
sns.kdeplot(
    x=df['Start_Lng'],
    y=df['Start_Lat'],
    cmap='Reds',
    fill=True,
    thresh=0.05
)
plt.title('Accident Density Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
